```python
import pandas as pd

employees = pd.read_csv("C:\\Users\\jouin\\Downloads\\archive (3)\\employees.csv")
salaries = pd.read_csv("C:\\Users\\jouin\\Downloads\\archive (3)\\salaries.csv")
titles = pd.read_csv("C:\\Users\\jouin\\Downloads\\archive (3)\\titles.csv")

employees.head()
```

```python
df_merged = titles.merge(salaries, on="emp_no")
```

```python
df_merged.groupby("title")["salary"].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

stipendio_per_titolo = df_merged.groupby("title")["salary"].mean().sort_values(ascending=False)

stipendio_per_titolo.plot(kind="bar")
plt.title("Stipendio medio per titolo")
plt.xlabel("Titolo")
plt.ylabel("Stipendio medio")
plt.show()
```
